```python
import pandas as pd

# Read the CSV file
orders_df = pd.read_csv("Orders_placed.csv")

# Define selected locations and products
selected_locations = ['California', 'New Jersey', 'Florida']
selected_products = ['Books', 'Jewelery', 'Toys', 'Shoes']

# Filter the DataFrame based on selected locations and products
orders_df = orders_df[(orders_df['location'].isin(selected_locations)) & (orders_df['product'].isin(selected_products))]
```

```python
orders_df.head()
```

```python
business_df = pd.read_csv("Business_registrations_data.csv")

# Define selected locations and products
selected_locations = ['California', 'New Jersey', 'Florida']
selected_products = ['Books', 'Jewelery', 'Toys', 'Shoes']

# Filter the DataFrame based on selected locations and products
business_df = business_df[(business_df['location'].isin(selected_locations)) & (business_df['product'].isin(selected_products))]
```

```python
business_df.head()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

# Load datasets
orders = orders_df
businesses = business_df
# Preprocess data
le = LabelEncoder()
orders['product'] = le.fit_transform(orders['product'])
businesses['product'] = le.fit_transform(businesses['product'])

# Cluster orders based on product
kmeans_orders = KMeans(n_clusters=5, random_state=42)
orders['cluster'] = kmeans_orders.fit_predict(orders[['product']])

# Cluster businesses based on product
kmeans_businesses = KMeans(n_clusters=5, random_state=42)
businesses['cluster'] = kmeans_businesses.fit_predict(businesses[['product']])

# Order matching algorithm
matched_orders = []
for _, order in orders.iterrows():
    matched_businesses = []
    for _, business in businesses.iterrows():
        if order['cluster'] == business['cluster'] and order['location'] == business['location']:
            if business['price_per_commodity'] <= order['max_affordable_price'] and business['max_capacity'] >= order['Order_size']:
                matched_businesses.append(business['Business_id'])
    if matched_businesses:
        matched_orders.append((order['Order_id'], matched_businesses))

# Save matched orders to a new dataframe
matched_orders_df = pd.DataFrame(matched_orders, columns=['Order_id', 'Business_id'])
matched_orders_df.to_csv('matched_orders.csv', index=False)
```
